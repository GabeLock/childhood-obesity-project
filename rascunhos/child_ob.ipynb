```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.api.types import CategoricalDtype
```

```python
with open('child_ob_gender.csv') as f:
    gdf = pd.read_csv(f)

with open('obesity_child_age.csv') as f:
    adf = pd.read_csv(f)
```

```python
adf.Time.unique()
```

```python
# adf['Time'] = adf.Time.astype("category", categories=adf.Time.unique(), ordered=True)
# gdf['Time'] = gdf.Time.astype("category", categories=gdf.Time.unique(), ordered=True)
time_cat_type = CategoricalDtype(categories=adf.Time.unique(), ordered=True)
adf['Time'] = adf['Time'].astype(time_cat_type)
gdf['Time'] = gdf['Time'].astype(time_cat_type)
```

```python
test = gdf[(gdf.Gender == 'F') | (gdf.Gender == 'M')]
```

```python
sns.pointplot(x="Time", y='PercentOB', hue='Gender', data=test)
plt.xticks(rotation=75, size=14)
plt.yticks(size=14)
plt.subplots_adjust(bottom=0.20)
plt.ylabel('Percent Obese', size=16, style='oblique')
plt.xlabel('')
plt.title('Historical US Childhood Obesity Rates (Gender)', size=18)
plt.show()
```

```python
test = adf[adf.Gender == 'All']
```

```python
sns.pointplot(x="Time", y='PercentObese', hue='Age', data=test)
plt.xticks(rotation=75, size=14)
plt.yticks(size=14)
plt.subplots_adjust(bottom=0.20)
plt.ylabel('Percent Obese', size=16, style='oblique')
plt.xlabel('')
plt.title('Historical US Childhood Obesity Rates (Age)', size=18)
plt.show()
```
